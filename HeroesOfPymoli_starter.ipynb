## Gender Demographics

```python
# Total_Count = purchase_data["Item ID"].count()
# Total_Count
#purchase_data = purchase_data.groupby("SN")
# count = purchase_data.groupby(['Gender']).count()
# count
#count.reset_index()
#purchase_data.reset_index()
#summary_table
#----------Array for Gender Type
# test = purchase_data["Gender"].unique()
# test
#-------------
MALE_INFO = purchase_data.loc[purchase_data["Gender"] == "Male",:]
#MALE_INFO
Male_Count = len(MALE_INFO["SN"].unique())
Percent_Male = (Male_Count / count) * 100
#Percent_Male

FMALE_INFO = purchase_data.loc[purchase_data["Gender"] == "Female",:]
#FMALE_INFO
Female_Count = len(FMALE_INFO["SN"].unique())
Percent_Female = (Female_Count / count) * 100
#Percent_Female

Other_INFO = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed",:]
#Other_INFO
Other_Count = len(Other_INFO["SN"].unique())
#Other_Count
Percent_Other = (Other_Count / count) * 100
#Percent_Other

Gender_demographics = pd.DataFrame({'Total Count': [Male_Count, Female_Count, Other_Count], 'Percentage of Players': [Percent_Male, Percent_Female, Percent_Other]},index=['Male', 'Female', 'Other / Non-Disclosed'])
#Gender_demographics
Gender_demographics["Percentage of Players"] = Gender_demographics["Percentage of Players"].map("{:.2f}".format)
Gender_demographics
```


## Purchasing Analysis (Gender)

* Create a summary data frame to hold the results

```python
Male_Purchase_Count = MALE_INFO["SN"].count()
Total_Male_Purchase_Price = (MALE_INFO["Price"].sum())
Average_Male_Purchase_Price = Total_Male_Purchase_Price / Male_Purchase_Count
Avg_Total_Male_Purchase_per_Person = Total_Male_Purchase_Price / Male_Count

Female_Purchase_Count = FMALE_INFO["SN"].count()
Total_Female_Purchase_Price = (FMALE_INFO["Price"].sum())
Average_Female_Purchase_Price = Total_Female_Purchase_Price / Female_Purchase_Count
Avg_Total_Female_Purchase_per_Person = Total_Female_Purchase_Price / Female_Count

Other_Purchase_Count = Other_INFO["SN"].count()
Total_Other_Purchase_Price = (Other_INFO["Price"].sum())
Average_Other_Purchase_Price = Total_Other_Purchase_Price / Other_Purchase_Count
Avg_Total_Other_Purchase_per_Person = Total_Other_Purchase_Price / Other_Count


Purchase_Analysis = pd.DataFrame({'Gender': ['MALE', 'FEMALE', 'Other'], 'Purchase Count': [Male_Purchase_Count, Female_Purchase_Count, Other_Purchase_Count], 'Average Purchase Price': [Average_Male_Purchase_Price, Average_Female_Purchase_Price, Average_Other_Purchase_Price], 'Total Purchase Value': [Total_Male_Purchase_Price, Total_Female_Purchase_Price, Total_Other_Purchase_Price], 'Avg Total Purchase per Person': [Avg_Total_Male_Purchase_per_Person, Avg_Total_Female_Purchase_per_Person, Avg_Total_Other_Purchase_per_Person]})
Purchase_Analysis["Average Purchase Price"] = Purchase_Analysis["Average Purchase Price"].map("${:.2f}".format)
Purchase_Analysis["Total Purchase Value"] = Purchase_Analysis["Total Purchase Value"].map("${:.2f}".format)
Purchase_Analysis["Avg Total Purchase per Person"] = Purchase_Analysis["Avg Total Purchase per Person"].map("${:.2f}".format)
Purchase_Analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#purchase_data
PC_PD = purchase_data["SN"].value_counts()

grouped_purchase_data = purchase_data.groupby(['SN'])

Total_Purchase_Value = grouped_purchase_data["Price"].sum()

Total_Purchase_Value.head()

Top_Spenders_summary_table = pd.DataFrame({"Purchase Count": PC_PD, "Average Purchase Price" : (Total_Purchase_Value / PC_PD) ,"Total Purchase Value": Total_Purchase_Value})
Top_Spenders_summary_table.head()

Top_Spenders_summary_table = Top_Spenders_summary_table.sort_values("Total Purchase Value", ascending=False)
Top_Spenders_summary_table["Average Purchase Price"] = Top_Spenders_summary_table["Average Purchase Price"].map("${:.2f}".format)
Top_Spenders_summary_table["Total Purchase Value"] = Top_Spenders_summary_table["Total Purchase Value"].map("${:.2f}".format)

Top_Spenders_summary_table.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

purchase_data_1 = purchase_data.loc[:, ['Item ID', 'Item Name', 'Price']]

grouped_item_data = purchase_data_1.groupby(['Item ID','Item Name']).sum()["Price"]
grouped_item_data
grouped_item_data_1 = purchase_data_1.groupby(['Item ID','Item Name']).mean()["Price"]

grouped_item_data_2 = purchase_data_1.groupby(['Item ID','Item Name']).count()["Price"]

MPI_summary_table = pd.DataFrame({ "Purchase Count" : grouped_item_data_2, "Item Price" : grouped_item_data_1, "Total Purchase Value" : grouped_item_data})
MPI_summary_table2 = pd.DataFrame({ "Purchase Count" : grouped_item_data_2, "Item Price" : grouped_item_data_1, "Total Purchase Value" : grouped_item_data})


MPI_summary_table = MPI_summary_table.sort_values("Purchase Count", ascending=False)
MPI_summary_table["Item Price"] = MPI_summary_table["Item Price"].map("${:.2f}".format)
MPI_summary_table["Total Purchase Value"] = MPI_summary_table["Total Purchase Value"].map("${:.2f}".format)
MPI_summary_table

MPI_summary_table.head()
```
